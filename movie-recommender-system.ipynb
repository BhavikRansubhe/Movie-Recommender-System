Here above it looks like dictionary , but it is string so we need to convert it into list by using ast.literal_eval<br>
So to take genres values of key name

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj): #to convert into list
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
# do same for keywords coulumn
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
#movies['cast'][0]  #take only top 3 name of cast as they are main leads

def convert3(obj):
    L=[]
    counter =0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter +=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
#movies['crew'][0]  #take only producers name

def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

Now here we need to make tags to recommend on that basis <br>
so Sam Worthington & Sam Mendes are different words having sam as common so we need to make that a single word like SamWorthington

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x]) #Eg Science Fiction -> ScienceFiction
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

supppose there are similar words like ['loved,'loving','love'] <br>
so we need common meaning like ['love','love','love'] <bt>Therefore we use nltk lib

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
#eg-
stem('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthington zoesaldana sigourneyweaver jamescameron')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')

# max_features tells how many max words you want 
# stop_words means we dont want words like 'a' 'the' 'ans' for removing common btwn tags 
```

```python
#To convert scipy sparse matrix into numpy array
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0] #vector of avtar movie
```

```python
cv.get_feature_names()  #here we can see top 5000 words which ocurred most frequent
```

Now lets calculate distance between 2 vectors :
we can use cosine similarity 

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
# to keep index in contact and to find similarity of first movie with other movies , hence with itself is 1
list(enumerate(similarity[0]))
```

```python
# To get top 5 similar movies 
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Iron Man')
```

```python
recommend('Avatar')
```
